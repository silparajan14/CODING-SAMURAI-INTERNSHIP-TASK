CREATE AN ITEM-BASED COLLABORATIVE FILTER ON THE ENTIRE DATASET

```python
# Recall this matrix that we created earlier of all movies and their user ID/ratings
userid_movietitle_matrix
```

```python
movie_correlations = userid_movietitle_matrix.corr(method = 'pearson', min_periods = 80)
movie_correlations
# pearson : standard correlation coefficient
# Obtain the correlations between all movies in the dataframe
```

```python
# Let's create our own dataframe with our own ratings!
myRatings = pd.read_csv("/content/My_Ratings.csv")
myRatings
```

```python
len(myRatings.index)
```

```python
myRatings['Movie Name'][0]
```

```python
similar_movies_list = pd.Series()

for i in range(len(myRatings)):
    similar_movie = movie_correlations[myRatings['Movie Name'][i]].dropna() # Get same movies with same ratings
    similar_movie = similar_movie.map(lambda x: x * myRatings['Ratings'][i]) # Scale the similarity by your given ratings
    similar_movies_list = pd.concat([similar_movies_list, similar_movie])

# Now, you can sort the similar_movies_list Series to get top recommendations
similar_movies_list = similar_movies_list.groupby(similar_movies_list.index).sum().sort_values(ascending=False)

# Print top recommendations
print(similar_movies_list.head(10))
```

```python

# Load the saved model
model_file_path = "/content/movie_correlations.pkl"
with open(model_file_path, 'rb') as f:
    loaded_model = pickle.load(f)

# Define a function to generate recommendations based on user inputs
def get_recommendations(movie1_rating, movie2_rating, movie1, movie2):
    print("Movie 1:", movie1)
    print("Movie 2:", movie2)

    # Check if movie names exist in the loaded model
    if movie1 not in loaded_model.columns:
        print("Movie 1 not found in the model.")
    if movie2 not in loaded_model.columns:
        print("Movie 2 not found in the model.")

    # Create a dictionary with the provided ratings
    movie_ratings = {movie1: movie1_rating, movie2: movie2_rating}
    print("Movie Ratings:", movie_ratings)

    similar_movies_list = pd.Series()
    for movie, rating in movie_ratings.items():
        similar_movie = loaded_model[movie].dropna() # Get similar movies
        similar_movie = similar_movie.map(lambda x: x * rating) # Scale similarity by rating
        similar_movies_list = pd.concat([similar_movies_list, similar_movie])
    similar_movies_list = similar_movies_list.groupby(similar_movies_list.index).sum().sort_values(ascending=False)
    recommended_movies = list(similar_movies_list.head(10).index)  # Return top 10 recommended movies

    print("Recommended Movies:", recommended_movies)  # Print recommended movies for debugging

    if not recommended_movies:
        return "No recommendations available."
    else:
        return ", ".join(recommended_movies)

# Define Gradio interface
interface = gr.Interface(
    fn=get_recommendations,
    inputs=[
        gr.Slider(minimum=0, maximum=5, step=0.1, label="Rating for Movie 1"),
        gr.Slider(minimum=0, maximum=5, step=0.1, label="Rating for Movie 2"),
        gr.Textbox(label="Name of Movie 1"),
        gr.Textbox(label="Name of Movie 2")
    ],
    outputs=gr.Textbox(label="Recommended Movies"),
    title="Movie Recommendation System"
)
interface.launch()
```
